```python
def grade_setter(x: int) -> str:
    if x == 0:
        return "Undefined grade"
    if x < 18:
        return "F" 
    if 19 <= x <38:
        return "E"
    if 39 <= x < 58:
        return "D" 
    if 58 <= x < 78:
        return "C" 
    if 78 <= x < 93:
        return "B" 
    if x>=93:
        return "A" 
```

# Fetching data

```python
url = 'https://raw.githubusercontent.com/DAVE3625-22H/Lab-1/main/students.csv'
df = pd.read_csv(url)

```

# Exploring data

```python
df
```

# Data Cleaning

```python
# Replaces all places where NaN=True.
df=df.replace(r'^\s*$', np.nan, regex=True)

# Drops NaN-values.
df.dropna(inplace=True)
df.isna().sum()
```

```python
# Could also change NaN to zero-value, and get undefined grade
df = df.fillna(0)

# Checks if there are any columns containing NaN-values.
df.isna().sum()
```

```python
df.info()
```

```python
# Removing outliers
col = ["FinalGrade"]

q1 = df[col].quantile(0.25)
q3 = df[col].quantile(0.75)
iqr = q3 - q1

df = df[~((df[col] < (q1 - 1.5* iqr)) | (df[col] > (q3 + 1.5*iqr))).any(axis=1)]
```

# Feature Engineering - Making letter grade

```python
# Applying the lambda function on 'Lettergrade'-column
df["LetterGrade"] = df["FinalGrade"].apply(lambda x: grade_setter(x))
```
